# Handling Outliers

```python
a = df.quantile(q=[0.75,0.25])
IQR = a.iloc[0] - a.iloc[1]
IQR
```

```python
upper = a.iloc[0]+(1.5*IQR)
lower = a.iloc[0]-(1.5*IQR)
```

```python
upper
```

```python
lower
```

```python
df['Length'] = np.where(df['Length'] < 0.36750, df['Length'].mean(), df['Length'])
df['Length'] = np.where(df['Length'] > 0.86250, df['Length'].mean(), df['Length'])
```

```python
sns.boxplot(df['Length'])
```

# Dependent and Independent Variable

```python
x = df.drop(columns='Rings')
y = df['Rings']
```

```python
x
```

```python
y
```

# Scaling

```python
from sklearn.preprocessing import scale
y = scale(y)
y
```

# Splitting data into training and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

# MLR

```python
from sklearn.linear_model import LinearRegression
MLR = LinearRegression()
MLR.fit(x_train, y_train)
```

```python
pred = MLR.predict(x_test)
pred
```

```python
MLR.predict([[2,0.45, 0.36, 0.09, 0.51, 0.22, 0.101, 0.150]])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, MLR.predict(x_test))
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,MLR.predict(x_test)))
```
